```python
import yfinance as yf 
import pandas as pd 
from pandas import Series,DataFrame
import numpy as np

import datetime
#VIS  imports
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

#define csv output
output_data = ('../df_csv/as_df.csv')
```

```python
# Create list of all companies we want records for. 
AS_Stock_list = ['NSANY', 'HMC', 'TM', 'HYMTF', 'MZDAY' ]

# Set Range we want to records from. 
start = datetime.date(2019,12,2) 
end = datetime.date(2020,12,3)


# Loop through records for all ticker symbols in our list.
for stock in AS_Stock_list:
     globals()[stock] = yf.download(stock,start,end)
```

```python
# Add Company columns to all data frames for merge.
NSANY['Company'] = 'Nissan'
HMC['Company'] = 'Honda'
TM['Company'] = 'Toyota'
HYMTF['Company'] = 'Hyundai'
MZDAY['Company'] = 'Mazda'


df_list = [NSANY, HMC, TM, HYMTF, MZDAY]

as_df = pd.concat(df_list)

as_df_reordered = ['Company', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

final_as_df = as_df[as_df_reordered]
final_as_df
```

```python
averages = [10,25,50]
for avg in averages:
    label = "MAvg for %s days" %(str(avg))
    NSANY[label] = NSANY['Adj Close'].rolling(window = avg, center = False).mean()
    HMC[label] = HMC['Adj Close'].rolling(window = avg, center = False).mean()
    TM[label] = TM['Adj Close'].rolling(window = avg, center = False).mean()
    HYMTF[label] = HYMTF['Adj Close'].rolling(window = avg, center = False).mean()
    MZDAY[label] = MZDAY['Adj Close'].rolling(window = avg, center = False).mean()
```

```python
fig = plt.figure(figsize=(15,10))
N=NSANY[['Adj Close', 'MAvg for 10 days', 'MAvg for 25 days', 'MAvg for 50 days']]
H=HMC[['Adj Close', 'MAvg for 10 days', 'MAvg for 25 days', 'MAvg for 50 days']]
T=TM[['Adj Close', 'MAvg for 10 days', 'MAvg for 25 days', 'MAvg for 50 days']]
HY=HYMTF[['Adj Close', 'MAvg for 10 days', 'MAvg for 25 days', 'MAvg for 50 days']]
M=MZDAY[['Adj Close', 'MAvg for 10 days', 'MAvg for 25 days', 'MAvg for 50 days']]
sub1 = fig.add_subplot(221) 
sub1.set_title('Nissan') 
sub1.plot(N)
sub1.legend(['Adj Close'])
sub1.legend(['MAvg for 10 days'])
sub1.legend(['MAvg for 25 days'])
sub1.legend(['MAvg for 50 days'])
sub2 = fig.add_subplot(222)
sub2.set_title('Honda')
sub2.plot(H)
sub2.legend(['Adj Close'])
sub2.legend(['MAvg for 10 days'])
sub2.legend(['MAvg for 25 days'])
sub2.legend(['MAvg for 50 days'])
sub3 = fig.add_subplot(223) 
sub3.set_title('Toyota') 
sub3.plot(T)
sub3.legend(['Adj Close'])
sub3.legend(['MAvg for 10 days'])
sub3.legend(['MAvg for 25 days'])
sub3.legend(['MAvg for 50 days'])
sub4 = fig.add_subplot(224)
sub4.set_title('Hyundai')
sub4.plot(HY)
sub4.legend(['Adj Close'])
sub4.legend(['MAvg for 10 days'])
sub4.legend(['MAvg for 25 days'])
sub4.legend(['MAvg for 50 days'])
#sub5 = fig.add_subplot(225)
#sub5.set_title('MZDAY')
#sub5.plot(M)
#sub5.legend(['Adj Close'])
#sub5.legend(['MAvg for 10 days'])
#sub5.legend(['MAvg for 25 days'])
#sub5.legend(['MAvg for 50 days'])
plt.tight_layout()
plt.show()
              
```

```python
NSANY['Daily Return'] = NSANY['Adj Close'].pct_change()
HMC['Daily Return'] = HMC['Adj Close'].pct_change()
TM['Daily Return'] = TM['Adj Close'].pct_change()
HYMTF['Daily Return'] = HYMTF['Adj Close'].pct_change()
MZDAY['Daily Return'] = MZDAY['Adj Close'].pct_change()
```
