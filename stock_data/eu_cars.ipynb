```python
import yfinance as yf 
import pandas as pd 
from pandas import Series,DataFrame
import numpy as np
from pylab import *
import datetime
#Visualisation imports
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
# Create list of all companies we want records for. 
EU_Stock_list = ['VWAGY', 'BAMXF', 'DMLRY', 'POAHY', 'VLVLY' ]

# Set Range we want to records from. 
start = datetime.date(2019,12,2) 
end = datetime.date(2020,12,3)


# Loop through records for all ticker symbols in our list.
for stock in EU_Stock_list:
     globals()[stock] = yf.download(stock,start,end)
```

```python
averages = [10,25,50]
for avg in averages:
    label = "MAvg for %s days" %(str(avg))
    VWAGY[label] = VWAGY['Adj Close'].rolling(window = avg, center = False).mean()
    BAMXF[label] = BAMXF['Adj Close'].rolling(window = avg, center = False).mean()
    POAHY[label] = POAHY['Adj Close'].rolling(window = avg, center = False).mean()
    VLVLY[label] = VLVLY['Adj Close'].rolling(window = avg, center = False).mean()
    DMLRY[label] = DMLRY['Adj Close'].rolling(window = avg, center = False).mean()
```

```python
fig = plt.figure(figsize=(15,10))
t=VWAGY[['Adj Close', 'MAvg for 10 days', 'MAvg for 25 days', 'MAvg for 50 days']]
u=VWAGY['Volume']
sub1 = fig.add_subplot(221) 
sub1.set_title('VW Adj Close & MAvg for 10, 25, and 50 days') 
sub1.plot(t)
sub1.legend(['Adj Close', 'MAvg for 10 days', 'MAvg for 25 days', 'MAvg for 50 days'])
sub2 = fig.add_subplot(222)
sub2.set_title('VW Volume')
sub2.plot(u)
sub2.legend(['Volume'])
plt.tight_layout()
plt.show()
```

```python
VWAGY['Daily Return'] = VWAGY['Adj Close'].pct_change()
BAMXF['Daily Return'] = BAMXF['Adj Close'].pct_change()
POAHY['Daily Return'] = POAHY['Adj Close'].pct_change()
VLVLY['Daily Return'] = VLVLY['Adj Close'].pct_change()
DMLRY['Daily Return'] = DMLRY['Adj Close'].pct_change()

sns.set(rc={"figure.figsize": (20, 10)}); np.random.seed(0)
a = VWAGY['Daily Return']
b=BAMXF['Daily Return']
c=POAHY['Daily Return'] 
d=VLVLY['Daily Return']
e=DMLRY['Daily Return']
```
