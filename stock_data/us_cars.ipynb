```python
import yfinance as yf 
import pandas as pd 
from pandas import Series,DataFrame
import numpy as np
from pylab import *
import datetime
#Visualisation imports
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

#define csv output
output_data = ('../df_csv/us_df.csv')
```

```python
# Create list of all companies we want records for. 
US_Stock_list = ['GM', 'F', 'FCAU', 'TSLA' ]

# Set Range we want to records from. 
start = datetime.date(2019,12,2) 
end = datetime.date(2020,12,3)


# Loop through records for all ticker symbols in our list.
for stock in US_Stock_list:
     globals()[stock] = yf.download(stock,start,end)
```

```python
averages = [10,25,50]
for avg in averages:
    label = "MAvg for %s days" %(str(avg))
    TSLA[label] = TSLA['Adj Close'].rolling(window = avg, center = False).mean()
    F[label] = F['Adj Close'].rolling(window = avg, center = False).mean()
    GM[label] = GM['Adj Close'].rolling(window = avg, center = False).mean()
    FCAU[label] = FCAU['Adj Close'].rolling(window = avg, center = False).mean()
```

We have used moving averages to even out the volatility on the daily stock movements and allow for a period analysis; the period chosen are short-term (10 days) medium-term (25 days) and long-term (50 days)

```python
fig = plt.figure(figsize=(15,10))
t=GM[['Adj Close', 'MAvg for 10 days', 'MAvg for 25 days', 'MAvg for 50 days']]
u=GM['Volume']
sub1 = fig.add_subplot(221) 
sub1.set_title('GM Adj Close & MAvg for 10, 25, and 50 days') 
sub1.plot(t)
sub1.legend(['Adj Close', 'MAvg for 10 days', 'MAvg for 25 days', 'MAvg for 50 days'])
sub2 = fig.add_subplot(222)
sub2.set_title('GM Volume')
sub2.plot(u)
sub2.legend(['Volume'])
plt.tight_layout()
plt.show()
```

General Motors shows little variablility this is evidenced by the all 3 periodic averages tracing the daily curve. while there was a significant drop in March due to the overall market covid-sell off the stock rebounded quickly showing steady growth and is up >10% from its pre-covid high.

```python
TSLA['Daily Return'] = TSLA['Adj Close'].pct_change()
F['Daily Return'] = F['Adj Close'].pct_change()
GM['Daily Return'] = GM['Adj Close'].pct_change()
FCAU['Daily Return'] = FCAU['Adj Close'].pct_change()
```

```python
stock_adjclose = yf.download(US_Stock_list,start,end)['Adj Close']
daily_return = stock_adjclose.pct_change()
```

```python
sns.set(rc={"figure.figsize": (18, 10)}); np.random.seed(0)
w = TSLA['Daily Return']
x=F['Daily Return']
y=GM['Daily Return'] 
z=FCAU['Daily Return']
subplot(2,2,1)
ax = sns.distplot(w.dropna(),bins=150,color='red')
plt.title("Tesla")
subplot(2,2,2)
ax = sns.distplot(x.dropna(),bins=150,color='red')
plt.title("Ford")
subplot(2,2,3)
ax = sns.distplot(y.dropna(),bins=150,color='red')
plt.title("GM")
subplot(2,2,4)
ax = sns.distplot(z.dropna(),bins=150,color='red')
plt.title("Chyrsler")
plt.show()
```

Tesla and Chrysler on average made you money; while GM was even and Ford lost you money most times.
